```python
import torch
import numpy as np
import normflows as nf

from matplotlib import pyplot as plt

from tqdm import tqdm
```

```python
torch.cuda.is_available()
```

```python
# input distribution
base = nf.distributions.base.DiagGaussian(2)

# Define list of flows
num_layers = 32
flows = []
# Here we define 32 layers
# In each layer, one input is used to compute the scale (s) and translation (t) for the other input
# In the next layer, we swap two inputs' roles otherwise one input stays invariant
for i in range(num_layers):
    # input has has size 1 corresponding to the input used to compute s and t
    # two hidden layers, each with 64 nodes
    # output has size 2, one output is the same as the input used to compute s and t, the other output is
    # the other input after been scaled and translated.
    param_map = nf.nets.MLP([1, 64, 64, 2], init_zeros=True)
    # They call this affine. Possibly because there is a translation?
    flows.append(nf.flows.AffineCouplingBlock(param_map))
    # Swap the roles of two inputs
    # The input used to compute s and t in the current layer will be scaled and translated according to the other input
    flows.append(nf.flows.Permute(2, mode='swap'))

# target distribution
target = nf.distributions.target.TwoMoons()
model = nf.NormalizingFlow(base, flows, target)
model = model.to('cuda')
device = 'cuda'
```

```python
# Plot target distribution
grid_size = 200
xx, yy = torch.meshgrid(torch.linspace(-3, 3, grid_size), torch.linspace(-3, 3, grid_size))
zz = torch.cat([xx.unsqueeze(2), yy.unsqueeze(2)], 2).view(-1, 2)
zz = zz.to(device)

log_prob = target.log_prob(zz).to('cpu').view(*xx.shape)
prob = torch.exp(log_prob)
prob[torch.isnan(prob)] = 0

plt.figure(figsize=(15, 15))
plt.pcolormesh(xx, yy, prob.data.numpy(), cmap='coolwarm')
plt.gca().set_aspect('equal', 'box')
plt.show()
```

```python
# Plot initial flow distribution
model.eval()
log_prob = model.log_prob(zz).to('cpu').view(*xx.shape)
model.train()
prob = torch.exp(log_prob)
prob[torch.isnan(prob)] = 0

plt.figure(figsize=(15, 15))
plt.pcolormesh(xx, yy, prob.data.numpy(), cmap='coolwarm')
plt.gca().set_aspect('equal', 'box')
plt.show()
```

```python
# Using Kullback Leibler divergence according to input distribution as the loss
max_iter = 1000

optimizer = torch.optim.Adam(model.parameters(), lr=5e-4, weight_decay=0.01)

for it in tqdm(range(max_iter)):
    optimizer.zero_grad()
    # My guess: here we generate 512 samples from the target distribution and transform them back
    # using the inverse of the model (very easy thank to the design of the coupled layer). 
    # Then, we compare these transformed samples with the input distribution and compute the loss until 
    # we obtain the optimal transformation model
    loss = model.reverse_kld(num_samples=512)
    if ~(torch.isnan(loss) | torch.isinf(loss)):
        loss.backward()
        optimizer.step()
```

```python
# Plot target distribution
f, ax = plt.subplots(1, 2, sharey=True, figsize=(15, 7))

log_prob = target.log_prob(zz).to('cpu').view(*xx.shape)
prob = torch.exp(log_prob)
prob[torch.isnan(prob)] = 0

ax[0].pcolormesh(xx, yy, prob.data.numpy(), cmap='coolwarm')

ax[0].set_aspect('equal', 'box')
ax[0].set_axis_off()
ax[0].set_title('Target', fontsize=24)

# Plot learned distribution
model.eval()
log_prob = model.log_prob(zz).to('cpu').view(*xx.shape)
model.train()
prob = torch.exp(log_prob)
prob[torch.isnan(prob)] = 0

ax[1].pcolormesh(xx, yy, prob.data.numpy(), cmap='coolwarm')

ax[1].set_aspect('equal', 'box')
ax[1].set_axis_off()
ax[1].set_title('Real NVP', fontsize=24)

plt.subplots_adjust(wspace=0.1)

plt.show()
```
